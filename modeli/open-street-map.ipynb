```python
import requests
import json
import time


overpass_url = "http://overpass-api.de/api/interpreter"

garage = [(60.10, 25.40)]

def overpass_query(latitude, longitude, around):
    overpass_query = "[out:json];(node(around:{},{},{})['shop'='car_repair'];);out;"\
    .format(around, latitude, longitude)

    while True:
        try:
            response = requests.get(overpass_url, params={'data': overpass_query})
            garages_around = response.json()
            return garages_around
        except:
            print("Try again...")
```

```python
%%time

around = int(80000)
garage_latlon_to_garages_in_80km_radius = {}

for i, garage in enumerate(garage):
    print("Start: ", i)
    
    latitude = float(garage[0])
    longitude = float(garage[1])
    print(latitude, longitude)
    
    garages_around = overpass_query(latitude, longitude, around)
    garage_latlon_to_garages_in_80km_radius[garage] = garages_around    
    print("End: ", i)
    print("------------------------------")
    time.sleep(5)
```
