```python
import numpy as np
import math
```

```python
def smape_wiki_1(A, F):
    return abs(F-A) / (0.5* (abs(F) + abs(A)))
```

```python
def smape_wiki_3(A, F):
    return abs(F-A) / (A + F)
```

```python
def smape(A, F):
    return abs(F-A) / max(F, A)
```

```python
smape_wiki_1(100, 90)
```

```python
smape(100, 90)
```

```python
smape_wiki_1(100, 110)
```

```python
smape(100, 110)
```

```python
smape_wiki_3(100, 110)
```

```python
A = 100
F_min = 0
F_max = 200
s_wiki = []
s = []
s_wiki_3 = []

for F in range(F_min, F_max):
    s_wiki.append(smape_wiki_1(A, F))
    s.append(smape(A, F))
    s_wiki_3.append(smape_wiki_3(A, F))
```

```python
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(10, 10))
# plt.plot(range(F_min, F_max), s_wiki, label='s_wiki')
plt.plot(range(F_min, F_max), s, label='s')
plt.plot(range(F_min, F_max), s_wiki_3, label='s_wiki_3')


# plt.plot(range(F_min, F_max), np.repeat(1, 1000))
plt.legend()
```

```python
plt.figure(figsize=(10, 10))
# plt.plot(range(F_min, F_max), 1-np.array(s_wiki), label='1-s_wiki')
plt.plot(range(F_min, F_max), 1-np.array(s), label='1-s')
plt.plot(range(F_min, F_max), 1-np.array(s_wiki_3), label='s_wiki_3')
plt.legend()
```
